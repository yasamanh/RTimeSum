```python
##Generating RTS Clustering output topic.rts (or just topic) using input from topic.out.sort
#TODO: need to take the Rel > 0 and check with formatted file
import csv 
import pdb
import os
        
#import sys
#reload(sys)
#sys.setdefaultencoding("utf-8")

file_path="/Users/ynh3/Documents/RTS/data/"
    
for file in os.listdir(file_path + "sorted/"):
    if (file == ".DS_Store"):
        continue
    print file
    sorted_file = open(file_path + "sorted/"+ file, "r")
    final_file = open(file_path+"final/microblog"+ file +".final")

    #sorted_list = list(csv.reader(sorted_file, delimiter='\t'))
    sorted_list=[]
    for line in sorted_file:
        sorted_list.append(line.replace('\n','').split('\t'))
    
    #final_list = list(csv.reader(final_file, delimiter='\t'))
    final_list=[]
    for line in final_file:
        final_list.append(line.replace('\n','').split('\t'))
    
    index = -1
    count = 0
    with open(file_path + '/RTS_input/' + file, 'w') as f:
        for i in range(0,len(final_list)):
            try:
                
                index = -1
                l = final_list[i][0].split(' ')
                if (int(l[7])>0):
                    if int(sorted_list[i][1])==int(l[4]):
                        #print i, sorted_list[i][1], l[4]
                        #pdb.set_trace()
                        index = i
                        count += 1
            
                        if index >= 0:
                            str = '\t'.join(x for x in ("0", file,"Test", sorted_list[index][1], "unknown undefined",sorted_list[index][2]))
                            f.write("{}\n".format(str))
                            
                    elif not int(sorted_list[i][1]) == int(l[4]):
                        #pdb.set_trace()
                        print "not found this id in sort file:", i, sorted_list[i][1], int(l[4])
            except IndexError:
                #pdb.set_trace()
                print i
                break
        f.close()
    print "number of tweets judged as relevant: ", count
```
