```python
import os
import collections
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

input_directory_path="/Users/ynh3/Documents/RTS/data/2016_clustering_input/"
output_directory_path="/Users/ynh3/Documents/RTS/data/2016_clustering_output/"
output_d_directory_path="/Users/ynh3/Documents/RTS/data/2016_clustering_output_detailed//"
    
for file in os.listdir(input_directory_path):
    if (file == ".DS_Store"):
        continue
    #print file
    
    f = open(input_directory_path+file,'r')
    lines = []
    for line in f:
     lines.append(line.replace('\n','').split('\t'))
    
    tweets = [e[2] for e in lines]
    ids = [e[1] for e in lines]
    print file, len(tweets), len(ids)
    
    true_k = 30
    #Convert a collection of raw documents to a matrix of TF-IDF features.
    vectorizer = TfidfVectorizer(stop_words='english')
    X = vectorizer.fit_transform(tweets)
    model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=10)
    model.fit(X)
    #print("Top terms per cluster:")
    order_centroids = model.cluster_centers_.argsort()[:, ::-1]
    terms = vectorizer.get_feature_names()
   
    clusters = collections.defaultdict(list)
    length = len(model.labels_)

    k = 0;
    for i in model.labels_:
        triple = str(k), ids[k], tweets[k]
        clusters[i].append(triple)  
        k += 1
 
        ## print out to file
    o_f = open(output_directory_path+file+".sorted","w")

    count = 0
    for clust in clusters :
      for i,t,d in clusters[clust]:
        s = '\t'.join(x for x in (i,t,d))
        o_f.write("{}\n".format(s))

    o_f.close()
    
    o_d_f = open(output_d_directory_path+file+".detailed","w")
    
    ## print out more details in file c
    for clust in clusters :    
        o_d_f.write("************************\n")
        s2 = "Cluster %d:" % clust, "Number of items: %d " %len(clusters[clust])
        o_d_f.write("{}\n".format(s2))                                                    
        #for ind in order_centroids[clust, :10]:
        #    o_d_f.write(terms[ind]).encode('utf8'))
        #o_d_f.write("\n")

        for i,t,d in clusters[clust]:
            s3 = ' '.join(x for x in(i,t,d))
            o_d_f.write(s3+"\n")
    o_d_f.close()
```
