```python
#using different ways of reading and parsing a file and counting the groupby
import matplotlib.pylab as plt
import pandas as pd
import csv
from operator import itemgetter
from itertools import groupby
```

```python
## Generate a group
cluster_file = open("/Users/ynh3/Documents/RTS/data/final_rts_results/clusters_assessors/37", "r")
preader = pd.read_csv(cluster_file, header = None, sep='\t')
df = pd.DataFrame(preader)
df.columns = ['A','B','C','D','E','F','G']
df_group = df.groupby('D')
size = df_group.size()

dic = dict(size)
sorted_cluster = sorted(dic.items(), key=itemgetter(1),reverse=True)

#How many itema and how much are clustered once and more than once
sum_all = sum(x[1] for x in sorted_cluster)
sum_one = sum(x[1] for x in sorted_cluster if x[1] == 1)
sum_others = sum(x[1] for x in sorted_cluster if x[1] > 1)
print sum_all, sum_one, sum_others
```
